```python
import numpy as np
import pandas as pd 
```

```python
train_data = pd.read_csv('../data/train.csv')
train_data
```

```python
test_data = pd.read_csv('../data/test.csv')
test_data
```

```python
# do some preprocessing!
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
train_data.isna().sum()
temp = ['Population', 'Hepatitis B','Polio', 'Diphtheria','Total expenditure','GDP', 'BMI','thinness  1-19 years', 'Alcohol','Schooling']
for col in temp:
    mean = train_data.groupby('Country')[col].transform('mean')
    # train_data[col].fillna(mean, inplace=True)
    # test_data[col].fillna(mean, inplace=True)
    train_data[col] = train_data[col].fillna(mean)
    test_data[col] = test_data[col].fillna(mean)
for col in temp:
    median = train_data[col].median()
    train_data[col] = train_data[col].fillna(median)
    test_data[col] = test_data[col].fillna(median)

x_train = train_data.drop(['Country', 'Status', 'Life expectancy'], axis=1)
y_train = train_data['Life expectancy']
x_test = test_data.drop(['Country', 'Status'], axis=1)

std_scaler = StandardScaler()
x_train = std_scaler.fit_transform(x_train)
x_test = std_scaler.transform(x_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_transformer = PolynomialFeatures(degree = 2)
x_train_poly = poly_transformer.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
# evaluate your model
from sklearn.metrics import r2_score
train_predictions = model.predict(x_train_poly)
r2_score(y_train, train_predictions)
```

```python
x_test_poly = poly_transformer.transform(x_test)
test_predictions = model.predict(x_test_poly)
submission = pd.DataFrame(test_predictions)
submission.columns = ['Life expectancy']
submission
```
