```python
import numpy as np
import pandas as pd 
```

```python
train_data = pd.read_csv('../data/bank_customers_train.csv')
test_data = pd.read_csv('../data/bank_customers_test.csv')
test_data
```

```python
# do some preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

train_data.dropna(subset=['job','marital'], inplace=True)

test_data['job'] = test_data['job'].fillna(test_data['job'].mode()[0])
test_data['marital'] = test_data['marital'].fillna(test_data['marital'].mode()[0])

m = ['education','default','housing','loan']
for i in m:
    train_data[i] = train_data[i].fillna(train_data[i].mode()[0])
    test_data[i] = test_data[i].fillna(test_data[i].mode()[0])
```

```python
train_data['marital'] =  LabelEncoder().fit_transform(train_data['marital'])
train_data['job'] =  LabelEncoder().fit_transform(train_data['job'])
test_data['marital'] =  LabelEncoder().fit_transform(test_data['marital'])  
test_data['job'] =  LabelEncoder().fit_transform(test_data['job'])  
```

```python
scaler = MinMaxScaler()

numeric = ['pdays','emp.var.rate','cons.price.idx','cons.conf.idx','nr.employed','euribor3m']

train_data[numeric] = scaler.fit_transform(train_data[numeric])
test_data[numeric] = scaler.transform(test_data[numeric])
```

```python
binary_cols = ['default','housing','loan'] 
train_data[binary_cols] = train_data[binary_cols].replace({'yes':1,'no':0})
train_data['contact'] = train_data['contact'].replace({'cellular':1,'telephone':0})
test_data[binary_cols] = test_data[binary_cols].replace({'yes':1,'no':0})
test_data['contact'] = test_data['contact'].replace({'cellular':1,'telephone':0})
```

```python
eduMap = {
    'illiterate' : 0,
    'basic.4y' : 1,
    'basic.6y' : 2,
    'basic.9y' : 3,
    'high.school': 4,
    'professional.course': 5,
    'university.degree': 6
}
weekdaysMap = {
    'mon' : 0,
    'tue' : 1,
    'wed' : 2,
    'thu' : 3,
    'fri' : 4,
    'sat' : 5,
    'sun' : 6
}
monthMap = {
    'feb' : 0,
    'jan' : 1,
    'mar' : 2,
    'apr' : 3,
    'may' : 4,
    'jun' : 5,
    'jul' : 6,
    'aug' : 7,
    'sep' : 8,
    'oct' : 9,
    'nov' : 10,
    'dec' : 11
}
poutcomeMap = {
    'failure' : -1,
    'nonexistent' : 0,
    'success' : 1
}
train_data['month'] = train_data['month'].map(monthMap)
train_data['education'] = train_data['education'].map(eduMap)
train_data['day_of_week'] = train_data['day_of_week'].map(weekdaysMap)
train_data['poutcome'] = train_data['poutcome'].map(poutcomeMap)
test_data['month'] = test_data['month'].map(monthMap)
test_data['education'] = test_data['education'].map(eduMap)
test_data['day_of_week'] = test_data['day_of_week'].map(weekdaysMap)
test_data['poutcome'] = test_data['poutcome'].map(poutcomeMap)
train_data
```

```python
x = train_data.drop(columns = ['y'])
y = train_data['y']
x_train , x_test , y_train ,y_test = train_test_split( x , y , test_size = 0.2)
```

```python
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.tree import DecisionTreeClassifier
model = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=4, min_samples_split=5, random_state=42), n_estimators=100, learning_rate=0.1, random_state=42)

model.fit(x_train,y_train)
```

```python
# evaluate model
from sklearn.metrics import f1_score
y_pred = model.predict(x_test)
print(f1_score(y_test,y_pred,pos_label = 'yes'))
```

```python
# predict test samples
submission = model.predict(test_data)
submission = pd.DataFrame(submission)
submission = submission.replace({'yes':1,'no':0})
submission.columns =['class']
submission
```
